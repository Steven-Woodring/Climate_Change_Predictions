```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
temps_path = "annual_global_land_ocean_temps.csv"
temps_df = pd.read_csv(temps_path)
temps_df
```

```python
#temps_df["Rolling Anomaly"] = temps_df["Anomaly"].rolling(10).mean()
#temps_df
```

```python
emissions_path = "CO2_emissions_1990_2018_cleaned.csv"
emissions_df = pd.read_csv(emissions_path)
emissions_df
```

```python
global_emissions = list(emissions_df["World"])
global_emissions
```

```python
temps_1990_2018 = temps_df[(temps_df["Year"] >= 1990) & (temps_df["Year"] <= 2018)]
temps_1990_2018["Emissions"] = global_emissions
temps_1990_2018
```

```python
#plot_df = temps_1990_2018.set_index("Emissions")

temps_1990_2018.plot.scatter("Emissions", "Temperature (Celsius)")
```

```python
X = temps_1990_2018['Emissions'].values.reshape(-1, 1)
X.shape
```

```python
y = temps_1990_2018['Temperature (Celsius)']
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
y_pred = model.predict(X)
print(y_pred.shape)
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.show()
```

```python
print(model.coef_)
print(model.intercept_)
print(model.score(X, y))
```
